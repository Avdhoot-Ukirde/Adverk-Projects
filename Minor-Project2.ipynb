```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

1. Face Recognition 

```python
img = cv2.imread(r"C:\Users\avdhoot\OneDrive\Desktop\Virtual-Environment-streamlit-Prediction-model\Computer-Vision-Project\p2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # convert BGR → RGB

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

model = cv2.CascadeClassifier(r"C:\Users\avdhoot\OneDrive\Desktop\Virtual-Environment-streamlit-Prediction-model\Computer-Vision-Project\haarcascade_frontalface_default.xml")
faces = model.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=9)


for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)


plt.imshow(img)
plt.axis("off")
plt.show()
```
